## 4-8. folium 기본 사용법

> - 이번 시간에는 **folium 을 사용한 기본적인 조작** 을 학습합니다.

```python
# 라이브러리 불러오기
import folium
import pandas as pd
import matplotlib.pyplot as plt
```

> ###  **4-8-1. 위도 경도 나타내기**

> - [구글맵](https://www.google.com/maps/) 에서 위도와 경도를 확인할 수 있습니다.
> - 원하는 위치를 검색 후, **위치를 우클릭하면 위도와 경도를 얻을 수 있습니다.**
> - folium.Map.save() 를 통해서 html 파일로 저장 가능합니다.

```python
# Map 함수를 이용해 지도를 focusing 합니다.
folium.Map(location=[37.50327278745055, 127.04160945554442], zoom_start=18)
```

```python
# 지도 사이즈 조절
f = folium.Figure(width=1000, height=500)
folium.Map(location=[37.50327278745055, 127.04160945554442], zoom_start=18).add_to(f)
```

```python
# 지도 저장
m = folium.Map(location=[37.50327278745055, 127.04160945554442], zoom_start=18)
m.save('myfolium.html') 
```

> ###  **4-8-2. 지도에 마커 표시하기**

```python
marker = folium.Marker([37.50327278745055, 127.04160945554442],
                       popup = '패스트 캠퍼스',     # 마커 이름 입력
                       icon = folium.Icon(color='blue'))
marker.add_to(m)
```

```python
f = folium.Figure(width=1000, height=500)
m.add_to(f)
```

```python
# 원 형태 마커 표시
marker = folium.CircleMarker([37.50327278745055, 127.04160945554442],
                             radius = 100,    # 범위
                             color = 'skyblue',
                             fill_color = 'skyblue',
                             popup = '패스트 캠퍼스',     # 마커 이름 입력
                             icon = folium.Icon(color='blue'))
marker.add_to(m)
```

```python
f = folium.Figure(width=1000, height=500)
m.add_to(f)
```

> ###  **4-8-3. 지도의 스타일 옵션**
> [folium 지도 스타일](https://deparkes.co.uk/2016/06/10/folium-map-tiles/)

```python
# tiles 옵션을 통해 지도 스타일을 바꿀 수 있습니다.
m = folium.Map(location=[37.50327278745055, 127.04160945554442], zoom_start=18, tiles='Stamen Terrain')

f = folium.Figure(width=1000, height=500)
m.add_to(f)
```

```python
# tiles 옵션을 통해 지도 스타일을 바꿀 수 있습니다.
m = folium.Map(location=[37.50327278745055, 127.04160945554442], zoom_start=18, tiles='Stamentoner')

f = folium.Figure(width=1000, height=500)
m.add_to(f)
```
