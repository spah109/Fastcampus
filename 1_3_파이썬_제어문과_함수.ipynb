> ### 2-3-1. **조건문**
> #### 조건문이란?
- 프로그래밍에서 조건에 따라 참과 거짓을 판단하는 문장입니다
- 일반적으로 **`비교 연산자`**를 기준으로 참/거짓을 판단하고, **`조건문`** 안에 있는 **`동작문`**을 순차적으로 실행합니다

> #### 비교 연산자
> - **값끼리 비교**해 참/거짓을 판단할때 사용하는 연산자
> - Python 에서 지원하는 비교 연산자는 다음과 같습니다
>
> |비교 연산자|설명|
|:---------|:---|
|A > B|A 가 B 보다 크다|
|A < B|A 가 B 보다 작다|
|A >= B|A 가 B 보다 크거나 같다|
|A <= B|A 가 B 보다 작거나 같다|
|A == B|A 와 B 가 같다|
|A != B&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;|A 가 B 와 같지 않다|

```python
# 코드를 입력하세요
a = 1
b = 2
a > b
```

```python
a == b
```

> #### 논리 연산자
> - **논리 연산을 이용**해 참/거짓을 판단할때 사용하는 연산자
> - Python 에서 지원하는 논리 연산자는 다음과 같습니다
>
> |논리 연산자|설명|
|:---------|:---|
|A and B|A 와 B 모두 참이다|
|A or B|A 와 B 중 적어도 하나는 참이다|
|not A&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;|A 가 거짓일 때 참이다|

```python
# 코드를 입력하세요
a = 1
b = 2
(a>1) and (b>1)
```

```python
(a>1) or (b>1)
```

```python
not (a>1)
```

> #### 그 밖의 연산자
> - Python 에서 제공하는 `in`, `not in` 을 활용해 참/거짓을 판단할때 사용하는 연산자
> - Python 에서 지원하는 논리 연산자는 다음과 같습니다
>
> |in 연산자|not in 연산자|
|:---------|:---|
|A in 리스트|A not in 리스트|
|A in 튜플|A not in 튜플|
|A in 문자열&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;|A not in 문자열|

```python
# 코드를 입력하세요
a = [1,2,3]
1 in a
```

```python
b = 'python'
'o' in b
```

> #### if, else 문
> **"*잠이 오면 커피를 마시고, 그렇지 않으면 차를 마셔라*"**
<br> <br>
> 

```python
# 코드를 입력해 주세요
sleepy = True

# 조건문
if sleepy == True:
    print("커피를 마신다")  # 동작문 A

else:
    print("차를 마신다")    # 동작문 B
```

```python
# 코드를 입력해 주세요
status = ['happy', 'angry', 'sleepy']

# 조건문
if 'sleepy' in status:
    print("커피를 마신다")  # 동작문 A

else:
    print("차를 마신다")    # 동작문 B
```

> #### elif 문
> **"*잠이 오면 커피를 마시고, 잠이 오진 않지만 화가나면 차를 마시고, 잠도 안오고 화도 안나면 물을 마셔라*"**

```python
# 코드를 입력해 주세요
status = ['happy']

# 조건문
if 'sleepy' in status:
    print("커피를 마신다")  # 동작문 A

elif 'angry' in status:
    print("차를 마신다")    # 동작문 B

else:
    print("물을 마신다")    # 동작문 C
```

> ### 2-3-2. **반복문**
> #### 반복문이란?
- **조건문이 거짓이될 때 까지 반복을 명령**하는 구문입니다
- Python 에서는 **`for 문`** 을 통해 변수를 받아 반복하거나, **`while 문`** 을 통해 일정 조건까지 반복할 수 있습니다

> #### while 문
> **"*잠이 오지 않을 때 까지 커피를 마셔라*"**

> 

```python
# 코드를 입력해 주세요
sleepy = True
awake_point = 10    # 잠이 깨는 시점

num_coffee = 0

while sleepy==True:                    # 조건 1 : 잠이 오면, 아래 실행
    num_coffee = num_coffee + 1
    print('커피를 {} 잔 마셨습니다!'.format(num_coffee))

    if num_coffee >= awake_point:      # 조건 2 : 잠이 깨는 시점이 되면, 아래 실행
        sleepy = False
        print('더이상 잠이 오지 않습니다')
```

> **"*잠이 오지 않을 때 까지 커피를 마시지만, 커피가 떨어지면 멈춰라*"**

```python
# 코드를 입력해 주세요
sleepy = True
awake_point = 10    # 잠이 깨는 시점
max_coffee = 5 

num_coffee = 0

while sleepy==True:                    # 조건 1 : 잠이 오면, 아래 실행
    num_coffee = num_coffee + 1
    print('커피를 {} 잔 마셨습니다!'.format(num_coffee))

    if num_coffee >= awake_point:      # 조건 2 : 잠이 깨는 시점이 되면, 아래 실행
        sleepy = False
        print('더이상 잠이 오지 않습니다')

    if num_coffee >= max_coffee:       # 조건 3 : 커피가 떨어지면, 아래 실행
        print('커피가 다 떨어졌습니다')
        break
```

> **"*무한 루프*"**

```python
# 코드를 입력해 주세요
num_coffee = 0

while True:
    num_coffee = num_coffee + 1
    print('커피를 {} 잔 마셨습니다!'.format(num_coffee))
```

> #### for 문
> - Python 에서 **가장 직관적이고 많이** 쓰이는 반복문
> - 리스트나 튜플, 문자열의 **첫 번째 요소부터 마지막 요소까지 차례로 변수를 받아서** 동작문이 반복됩니다
> - 기본 문법
    - `for` 받아올 변수 `in` [변수1, 변수2, 변수3, ...]: 

> **"*커피, 차, 물을 차례로 마셔라*"**

> 

```python
# 코드를 입력해 주세요
drink = ['커피', '차', '물']

for item in drink:
    print('{} 을(를) 마셨습니다!'.format(item))
```

> 동작문 Skip 하기 (`continue`)

```python
# 동작문 Skip 하기
drink = ['커피', '차', '물']

for item in drink:
    if item=='차':
        continue

    print('{} 을(를) 마셨습니다!'.format(item))
```

> 반복문 중간에 나오기 (`break`)

```python
# 반복문 중간에 나오기
drink = ['커피', '차', '물']

for item in drink:
    print('{} 을(를) 마셨습니다!'.format(item))

    if item=='차':
        break
```

> `range` 함수 사용하기

```python
for i in range(10):
    print('커피를 {} 잔 마셨습니다!'.format(i))
```

```python
drink = ['커피', '차', '물']

for i in range(len(drink)):
    item = drink[i]
    print('{} 을(를) 마셨습니다!'.format(item))
```

> ### 2-3-3. **함수**
> #### 함수 (Function) 이란?
> 
>
> - **입력 값을 받아서 일정한 처리를 진행하고 출력값을 반환**하는 장치
> - **반복되는 작업을 묶어서 재사용** 할 수 있도록 해주기 때문에 매우 중요한 개념입니다
> - 기본 문법
    - `def` 함수명(입력인수): <br>
        `return` 출력인수



```python
# 코드를 입력해 주세요
def sum(a, b):
    return a + b
```

```python
a = 1
b = 2
sum(a, b)
```

> 함수를 사용했을 때

```python
# 코드를 입력해 주세요
def drink_beverage(drink):
    for item in drink:
        print('{} 을(를) 마셨습니다!'.format(item))
```

```python
drink_beverage(['커피', '차', '물'])
drink_beverage(['콜라', '사이다', '우유'])
```

> 출력값 활용하기

```python
# 출력이 하나일 경우
def drink_beverage(drink):
    for item in drink:
        print('{} 을(를) 마셨습니다!'.format(item))

    return len(drink)
```

```python
num_drink = drink_beverage(['커피', '차', '물'])
print('총 {} 잔 마셨습니다'.format(num_drink))
```

```python
# 출력이 여러개일 경우
def drink_beverage(drink):
    for item in drink:
        print('{} 을(를) 마셨습니다!'.format(item))

    return len(drink), drink[-1]
```

```python
num_drink, last_drink = drink_beverage(['커피', '차', '물'])
print('총 {} 잔 마셨습니다'.format(num_drink))
print('마지막으로 {} 을(를) 마셨습니다'.format(last_drink))
```

> 다중 입력값 사용하기

```python
# 코드를 입력해 주세요
def drink_beverage(drink1, drink2, drink3):
    drink = [drink1, drink2, drink3]
    for item in drink:
        print('{} 을(를) 마셨습니다!'.format(item))
```

```python
drink1 = '커피'
drink2 = '차'
drink3 = '물'

drink_beverage(drink1, drink2, drink3)
```
