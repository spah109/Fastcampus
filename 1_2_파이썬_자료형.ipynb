## 2-2. Python 자료형

> #### **변수와 자료형**
> - 변수는 **자료형의 값을 담는 그릇**
> - 변수는 Python 이 취급하는 다양한 자료형을 모두 담을 수 있습니다
    - Python 자료형은 **`숫자형(int, float)`**, **`문자열(str)`**, **`참/거짓(bool)`**, **`리스트(list)`**, **`튜플(tuple)`**, **`딕셔너리(dict)`**, **`집합(set)`** 이 있습니다
> - 변수의 이름은 일반적으로 **"알파벳 소문자 + 숫자(혹은 언더바 `_`)"** 형식을 사용합니다 (ex. `a`, `a1`, `var1`, `var_2`, 등)
    - 변수의 이름으로 적절하지 않은 조합
        - 알파벳 이외의 언어로 된 변수 (한글, 중국어, 일본어 등)
        - 숫자가 맨 앞에 오는 경우 (ex. `1a`)
        - 언더바 외 특수 문자 (ex. `*`, `/`, `&`, `$`, `#`, `~` 등)
        - 변수명 사이에 공백이 있는 경우

```python
# 코드를 입력해 주세요
a = 1
b = 'python'
c = [1,2,3]
```

```python
print(a, b, c)
```

```python
# 코드를 입력해 주세요
a, b, c = 1, 'python', [1,2,3]
```

```python
print(a, b, c)
```

> #### **논리형**
> - 참/거짓(Boolean) 형태의 자료형입니다
> - `True`, `False` 조건과, 아무것도 없다는 뜻인 `None` 타입이 있습니다
> - Python 에서

> #### **리스트형**
> - **요소(값)의 모음** 형태의 자료형입니다
> - 묶고 싶은 요소(값)들을 대괄호 (`[]`) 를 사용하여 감싸주어 만듭니다
    - 리스트명 = [요소1, 요소2, 요소3, ...]
> - Python 의 **모든 자료형**을 리스트의 요소로 담을수 있습니다




```python
# 코드를 입력해 주세요
a = [1,2,3]
type(a)
```

```python
a[0]
```

```python
a[0] + a[1]
```

```python
a[0:2]
```

```python
b = [a, 4]
b
```

```python
b[0][-1]
```

```python
# 리스트 더하기
a = [1,2,3]
b = [4,5,6]
a + b
```

```python
# 리스트 곱하기
a = [1,2,3]
a * 3
```

```python
# 리스트 값 바꾸기
a = [1,2,3]
a[0] = 4
a
```

```python
# 리스트 요소 추가
a = [1,2,3]
a.append(4)
a
```

```python
# 리스트 요소 제거
a.remove(4)
a
```

```python
# 리스트 요소 삽입
a = [1,2,3]
a.insert(1,4)
a
```

```python
# 리스트 정렬
a.sort(reverse=False)
a
```

```python
# 리스트 위치 반환
a = [1,2,3]
a.index(2)
```

```python
# 리스트 개수 세기
a = [1,1,2,3]
a.count(1)
```

```python
# 리스트 길이 파악
len(a)
```

> #### **튜플형**
> - 리스트형과 거의 비슷하지만, 변수를 **읽기 전용** 으로 만들고 싶을때 사용합니다
> - 튜플 값은 삭제, 수정이 불가능 합니다
> - 묶고 싶은 요소(값)들을 괄호 (`()`) 를 사용하여 감싸주어 만듭니다
    - 튜플명 = (요소1, 요소2, 요소3, ...)




```python
# 코드를 입력해 주세요
a = (1,2,3)
type(a)
```

```python
a.append(4)
```

```python
a.remove(3)
```

```python
a[0] = 4
```

> #### **딕셔너리형**
> - **키(Key): 값(Value) 쌍 모음** 사전 형태의 자료형입니다
> - 리스트나 튜플과 다르게 **순서가 없습니다**
> - 값을 찾고 싶을 때, **매우 빠르게 값을 찾을 수 있습니다**
> - 키와 값 쌍들을 중괄호 (`{}`) 를 사용하여 감싸주어 만듭니다
    - 딕셔너리명 = {'Key1': Value1, 'Key2': Value2, ...}
> - 키(Key) 는 문자형/튜플, 값(Value) 은 모든 자료형을 담을수 있습니다




```python
a = {'Bulldog': 'dog', 'Munchkin' : 'cat'}
type(a)
```

```python
# 키 검색
a.keys()
```

```python
# 값 검색
a['Bulldog']
```

```python
# 값 추가
a['Poodle'] = 'dog'
a
```

```python
# 값 변경
a['Bulldog'] = 'cat'
a
```

```python
# 요소 삭제
del a['Bulldog']
a
```

```python
# 모두 삭제
a.clear()
a
```

> #### **집합형**
> - **순서가 없고, 중복이 허용되지 않는** 자료형입니다
> - 묶고 싶은 요소들을 중괄호 (`{}`) 를 사용하여 감싸주어 만듭니다
    - 집합명 = {요소1, 요소2, 요소3, ...}




```python
a = {1,2,3}
type(a)
```

```python
a = {1,1,2,3}
a
```

```python
b = {'H', 'E', 'L', 'L', 'O'}
b
```

```python
# 교집합
a = {1,2,3,4,5}
b = {4,5,6,7,8}
a & b
```

```python
# 합집합
a = {1,2,3,4,5}
b = {4,5,6,7,8}
a | b
```

```python
# 차집합
a = {1,2,3,4,5}
b = {4,5,6,7,8}
a - b
```

```python
# 다른 자료형으로 변환
a = {1,2,3,4,5}
print(type(a))
a = list(a)
print(type(a))
```
